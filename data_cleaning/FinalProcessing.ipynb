```python
amazon_categories_path = "/Users/COMPUTER/PycharmProjects/pythonProject2/datasets/amazon_categories.csv"
amazon_categories = pd.read_csv(amazon_categories_path)
amazon_categories.head()
```

```python
amazon_with_category = pd.merge(amazon, amazon_categories, left_on='category_id', right_on='id', how='left')
amazon_with_category.head()
```

```python
amazon_with_category = amazon_with_category.drop(['category_id','id'], axis = 1)
amazon_with_category.head()
```

```python
#create unique identifier for all products
temu['UID'] = ['T' + str(i) for i in range(len(temu))]
walmart['UID'] = ['W' + str(i) for i in range(len(walmart))]
amazon_with_category['UID'] = ['A' + str(i) for i in range(len(amazon_with_category))]
shein['UID'] = ['S' + str(i) for i in range(len(shein))]
```

```python
temu.head()
```

```python
# Standardize and select necessary columns
temu_standardized = temu.rename(columns={'title': 'Title', 'price': 'Price'})[['UID', 'Title', 'Price']]
walmart_standardized = walmart.rename(columns={'Product Name': 'Title', 'Sale Price': 'Price'})[['UID', 'Title', 'Price']]
amazon_standardized = amazon_with_category.rename(columns={'title': 'Title', 'price': 'Price'})[['UID', 'Title', 'Price']]
shein_standardized = shein.rename(columns={'name': 'Title', 'price': 'Price'})[['UID', 'Title', 'Price']]
```

```python
temu_standardized.head()
```

```python
# Combine all dataframes
Main = pd.concat([temu_standardized, walmart_standardized, amazon_standardized, shein_standardized], ignore_index=True)
Main.head()
```

```python
amazon_with_category['URL'] = amazon_with_category['productURL']
shein['URL'] = shein['url']
walmart['URL'] = walmart['Product Url']
```

```python
shein.head()
```

```python
URL = pd.concat([
    amazon_with_category[['UID', 'URL']],
    shein[['UID', 'URL']],
    walmart[['UID', 'URL']],
], ignore_index=True)
URL.head()
```

```python
temu['Rating'] = temu['goods_score']
temu['Sales'] = temu['sales_volume']
amazon_with_category['Rating'] = amazon_with_category['stars']
amazon_with_category['Sales'] = amazon_with_category['boughtInLastMonth']

Rating_Sales = pd.concat([
    temu[['UID', 'Rating', 'Sales']],
    amazon_with_category[['UID', 'Rating', 'Sales']],
], ignore_index=True)
```

```python
amazon_with_category['Keywords'] = amazon_with_category['category_name']
walmart['Keywords'] = walmart['Category']
temu['Keywords'] = temu['leve_2_category_name']
shein['Keywords'] = shein['description']

Keywords = pd.concat([
    amazon_with_category[['UID', 'Keywords']],
    walmart[['UID', 'Keywords']],
    temu[['UID', 'Keywords']],
    shein[['UID', 'Keywords']],
], ignore_index=True)
```
